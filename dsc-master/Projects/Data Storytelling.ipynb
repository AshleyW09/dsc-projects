# Starbucks vs McDonald's

We are going to compare the nutrition of drinks at Starbucks and McDonald's to see which company has the healtier drinks.

First we are going to clean the data, so there is only drinks from the two comapnies.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
starbucks = pd.read_csv("starbucks_drinkMenu_expanded.csv")
mcdonalds = pd.read_csv("menu.csv")
```

```python
starbucks.head()
```

```python
starbucks = starbucks[starbucks['Beverage_prep']=='Tall']
```

```python
mcdonalds.head()
```

```python
mcdonalds = mcdonalds[mcdonalds['Category']=='Coffee & Tea']
```

```python
mcdonalds = mcdonalds[mcdonalds['Serving Size']=='12 fl oz cup']
```

Next we will find the top five drinks with the most calories at Starbucks.

```python
top_starbucks = starbucks.nlargest(5, 'Calories')
```

```python
fig = plt.figure(figsize = (10, 5))
 
plt.barh(top_starbucks['Beverage'], top_starbucks['Calories'])
 
plt.xlabel("Calories")
plt.title("Top 5 Starbucks Drinks with the Most Calories")
fig.autofmt_xdate()
plt.show()
```

Now we will find the top five drinks with the most calories at McDonald's.

```python
top_mcdonalds = mcdonalds.nlargest(5, 'Calories')
```

```python
fig = plt.figure(figsize = (10, 5))
 
plt.barh(top_mcdonalds['Item'], top_mcdonalds['Calories'])
 
plt.xlabel("Calories")
plt.title("Top 5 McDonald's Drinks with the Most Calories")
fig.autofmt_xdate()
plt.show()
```

We can see that the Caramal Apple Spice has the most calories at Starbucks. It has 210 calories. Then looking at the McDonald's drink with the most calories, we have the Frappe Chocolate Chip at 530 calories. All of McDonald's highest calorie drinks have more than 200 calories.

Next we are going to compare the sugar contents of drinks at Starbucks and McDonald's.

```python
starbucks_sugar = starbucks.nlargest(5, ' Sugars (g)')
```

```python
fig = plt.figure(figsize = (10, 5))
 
plt.barh(starbucks_sugar['Beverage'], starbucks_sugar[' Sugars (g)'])
 
plt.xlabel("Sugars")
plt.title("Top 5 Starbucks Drinks with the Most Sugar")
fig.autofmt_xdate()
plt.show()
```

```python
mcdonalds_sugar = mcdonalds.nlargest(5, 'Sugars')
```

```python
fig = plt.figure(figsize = (10, 5))
 
plt.barh(top_mcdonalds['Item'], top_mcdonalds['Sugars'])
 
plt.xlabel("Sugars")
plt.title("Top 5 McDonald's Drinks with the Most Sugar")
fig.autofmt_xdate()
plt.show()
```

We are seeing a similar trend to the calories in drinks. McDonald's drinks have higher amounts of sugar in them compared to the Starbucks drinks. The highest sugar content in a Starbucks drink is close to 50g of sugar and this is in the Caramel Apple Spice which also had the highest calories for a Starbucks drink. The McDonald's drinks had 3 drinks with more than 50g of sugar with the Frappe Chocoloate Chip being the drink with the most sugar.

## Conclusion

In conclusion, the Starbucks drinks have less calories and less sugar than the McDonald's drinks.
